```python
import pandas as pd
df=pd.read_csv('English.csv')
df.head()
```

```python
df=df.rename(columns={'English words/sentences':'words_sentences'})
df.head()
```

```python
# Replacing punctuations with space
df['words_sentences_processed'] = df['words_sentences'].str.replace("[^a-zA-Z0-9]", " ")
```

```python
df
```

```python
# make entire text lowercase
df['words_sentences_processed'] = [review.lower() for review in df['words_sentences_processed']]
```
